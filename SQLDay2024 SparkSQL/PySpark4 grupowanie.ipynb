## PySpark
### grupowanie

```python
# utworzenie dataframe
df = spark.sql("SELECT ProductID, Name, ProductNumber, MakeFlag, Color, Size, ListPrice FROM AdventureWorks.Production_Product")
```

```python
# zliczanie wierszy
df.count()
```

```python
df.groupBy("Color").count().show()
```

```python
df.groupBy("Color").sum("ListPrice").show()
df.groupBy("Color").min("ListPrice").show()
df.groupBy("Color").max("ListPrice").show()
```

```python
df.groupBy("Color", "Size").sum("ListPrice").show()
```
