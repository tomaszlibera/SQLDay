```python
%%sql

-- WIDOK - możemy się do niego odwoływać w każdej sesji
CREATE OR REPLACE VIEW BikesMTB 
AS 
SELECT ProductID, Name, Color FROM Production_Product WHERE ProductSubcategoryID = 1

SELECT * FROM BikesMTB LIMIT 5
```

```python
%%sql
-- TYMCZASOWY WIDOK - widoczny tylko dla bieżącego użytkownika

CREATE OR REPLACE TEMPORARY VIEW BikesRoad
AS 
SELECT ProductID, Name, Color FROM Production_Product WHERE ProductSubcategoryID = 2;

SELECT * FROM BikesRoad LIMIT 5
```

```python
%%sql
-- GLOBAL TEMPORARY VIEW - widoczny w ramach bieżącej sesji (klastra spark)

CREATE OR REPLACE GLOBAL TEMPORARY VIEW BikesTouring
AS 
SELECT ProductID, Name, Color FROM Production_Product WHERE ProductSubcategoryID = 3;

SELECT * FROM BikesTouring LIMIT 5
```
