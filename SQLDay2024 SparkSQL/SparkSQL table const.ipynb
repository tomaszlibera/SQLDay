```python
%%sql
DROP TABLE IF EXISTS Products;

-- tworzenie tabeli
-- wspierane 2 rodzaje ograniczeń: NULL/ NOT NULL oraz CHECK  
CREATE TABLE Products (
    ProductID int NOT NULL,
    Name string NOT NULL,
    Color string NOT NULL
) USING DELTA
```

```python
%%sql
-- wstawienie danych
INSERT INTO Products (ProductID, Name, Color)
SELECT ProductID, Name, Color FROM Production_Product WHERE Color IS NOT NULL LIMIT 5
```

```python
%%sql
-- próba wstawienia danych z NULL
INSERT INTO Products (ProductID, Name, Color)
SELECT ProductID, Name, Color FROM Production_Product WHERE Color IS NULL LIMIT 5
```

```python
%%sql 
-- zmiana NOT NULL na NULL

ALTER TABLE Products CHANGE COLUMN Color DROP NOT NULL
```

```python
%%sql
-- próba wstawienia danych z NULL #2
INSERT INTO Products (ProductID, Name, Color)
SELECT ProductID, Name, Color FROM Production_Product WHERE Color IS NULL LIMIT 5
```
