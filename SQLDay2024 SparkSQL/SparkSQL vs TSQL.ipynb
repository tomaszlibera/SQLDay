## SparkSQL vs TSQL
Przykłady funkcji: https://spark.apache.org/docs/2.3.1/api/sql/

 


```python
%%sql
-- sparkSQL to nie TSQL
SELECT TOP 10 * FROM Production_Product
```

```python
%%sql
-- kilka innych różnic
SELECT * FROM Production_Product LIMIT 10 
```

```python
%%sql
-- kilka innych różnic
SELECT * FROM Production_Product WHERE Color IS NULL LIMIT 10
```

```python
%%sql
-- NULL vs COALESCE
SELECT 
    ProductID, Name, Color
    -- ,ISNULL(Color, 'no color')
    ,COALESCE(Color, 'no color') 
FROM Production_Product 
WHERE Color IS NULL LIMIT 10
```

```python
%%sql
-- []
SELECT 
    ProductID
    -- ,Name AS [Product Name]
    ,Name AS `Product Name`
    ,Color
FROM Production_Product 
WHERE Color IS NULL LIMIT 10
```

```python
%%sql
-- GETDATE vs current_timestamp/ current_date

-- SELECT GETDATE()

SELECT current_timestamp(), current_date()
```

```python
%%sql

-- SELECT FORMAT(GETDATE(), 'yyyyMMdd'), CONVERT(varchar(8), GETDATE(), 112)

SELECT DATE_FORMAT(current_timestamp(), 'yyyyMMdd'), DATE_FORMAT(current_timestamp(), 'yyyy/MM/dd')
```

```python
%%sql
-- SELECT EOMONTH(GETDATE())
SELECT LAST_DAY(current_timestamp())
```

```python
%%sql
-- SELECT DATEADD(DAY, 7, GETDATE())
SELECT DATE_ADD(current_timestamp, 7), ADD_MONTHS(current_timestamp, 7)
```

```python
%%sql
-- SELECT DATEADD(DAY, 7, GETDATE())
SELECT DATE_ADD(current_timestamp, -7)
```

```python
%%sql
-- SELECT IIF(8%2=0, 'parzysta', 'nieparzysta')

SELECT if(8%2=0, 'parzysta', 'nieparzysta'), if(7%2=0, 'parzysta', 'nieparzysta')
```
