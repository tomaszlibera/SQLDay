## PySpark
### ramki danych, aliasy, sortowanie

```python
# utworzenie data frame
df = spark.sql("SELECT ProductID, Name, ProductNumber, MakeFlag, Color, ListPrice FROM AdventureWorks.Production_Product")
```

```python
# wyświetlenie danych
df.show()
```

```python
# wyświetlenie danych #2
display(df)
```

```python
# ograniczenie liczby wierszy
df.show(5)
```

```python
# ograniczenie liczby wyświetlanych wierszy
display(df.head(5))
```

```python
# wyświetlenie schematu ramki danych
df.printSchema()
```

```python
# wybieranie kolumn
df.select(df.ProductID, df.Name, df.Color, df.ListPrice)
```

```python
# wybrane kolumny - show
df.select(df.ProductID, df.Name, df.Color, df.ListPrice).show(5)
```

```python
# wybrane kolumny - display (+ ograniczenie liczby wierszy) 
display(df.select(df.ProductID, df.Name, df.Color, df.ListPrice).head(5))
```

```python
# aliasy
df.select(df.ProductID.alias("ID"), df.Name.alias("ProductName"), df.Color, df.ListPrice.alias("Price")).show(3)
```

```python
# nazwa kolumny = wyrażenie
df.select(df.ListPrice * 2).show(3)

# alias
df.select((df.ListPrice * 2).alias("Price2")).show(3)
```

```python
# sortowanie
df.sort(df.Color, df.Name).show(10)
```

```python
from pyspark.sql.functions import col
df.sort(col("Color"), col("Name")).show(5)
```

```python
# sortowanie malejąco
df.sort(df.Color.desc(), df.Name).show(10)
```

```python
# sortowanie i nie skracanie do 20 znaków

df.sort(df.Color.desc(), df.Name).show(10, truncate=False)
```

```python
# orderBy

df.orderBy("Name").show()
df.orderBy(df.Name).show()
df.orderBy(col("Name")).show()

# orderBy malejąco
df.orderBy(df.Name.desc()).show()
df.orderBy(col("Name").desc()).show()
```
