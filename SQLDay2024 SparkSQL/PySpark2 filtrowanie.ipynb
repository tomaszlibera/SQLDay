## PySpark
### Filtrowanie

```python
# utworzenie data frame
df = spark.sql("SELECT ProductID, Name, ProductNumber, MakeFlag, Color, ListPrice FROM AdventureWorks.Production_Product")
```

```python
# filter

df.filter(df.Color=='Blue').show(3)
df.filter(df.Color!='Blue').show(3)
```

```python
# filter
# bardziej SQL-owo

df.filter("Color=='Blue'").show(3)
df.filter("Color!='Blue'").show(3)
```

```python
# distinct - usuwanie dupikatów

df.select(df.Color).distinct().show()
```

```python
# filter

df.filter(df.MakeFlag==True).show(3)
df.filter(df.MakeFlag!=True).show(3)
```

```python
#filter, where, between

df.filter(df.ListPrice.between(100,500)).show(3)
df.where(df.ListPrice.between(100,500)).show(3)
```

```python
# filter, contains

df.filter(df.Name.contains("Ball")).show()
```

```python
# filter, startswith, endwith

df.filter(df.Name.startswith("B")).show()

df.filter(df.Name.endswith("karm")).show()
```

```python
# wyłączenie wrażliwości na wielkość liter w foltrowaniu

from pyspark.sql.functions import lower
from pyspark.sql.functions import col


df.filter(lower(col("Name")).startswith("b")).show()

```

```python
# isNull, isNotNull

df.filter(df.Color.isNull()).show(5)
df.filter(df.Color.isNotNull()).show(5)
```

```python
# when, alias dla większej czytelności

from pyspark.sql.functions import when

df.select(df.ProductID, 
          df.Name,
          df.ListPrice,
          when(df.ListPrice<100,"low").when(df.ListPrice>300,"high").otherwise("middle").alias("PriceInfo")
          ).show(100)
```

```python
%%sql
SELECT ProductID, Name, ProductNumber, MakeFlag, Color, ListPrice
FROM Production_Product 
WHERE Name LIKE '%Wheel' 
ORDER BY ListPrice DESC    
LIMIT 5
```

```python
df.filter(df.Name.endswith("Wheel")).sort(df.ListPrice.desc()).show(5, truncate=False)
```
