## PySpark
### łączenie tabel

```python
# utworzenie dataframe
df_prod = spark.sql("SELECT ProductID, Name, ProductNumber, MakeFlag, Color, Size, ListPrice, ProductSubcategoryID FROM AdventureWorks.Production_Product")

df_subcat = spark.sql("SELECT * FROM AdventureWorks.Production_ProductSubcategory")
df_cat = spark.sql("SELECT * FROM AdventureWorks.Production_ProductCategory")
```

```python
# inner join
df_prod.join(df_subcat,df_prod.ProductSubcategoryID == df_subcat.ProductSubcategoryID).show(5)

df_prod.join(df_subcat,df_prod.ProductSubcategoryID == df_subcat.ProductSubcategoryID).count()
```

```python
# inner join
df_prod.join(df_subcat,df_prod.ProductSubcategoryID == df_subcat.ProductSubcategoryID, "inner").show(5)

df_prod.join(df_subcat,df_prod.ProductSubcategoryID == df_subcat.ProductSubcategoryID, "inner").count()
```

```python
# outer join (full join)
df_prod.join(df_subcat,df_prod.ProductSubcategoryID==df_subcat.ProductSubcategoryID,"outer").show(5)

df_prod.join(df_subcat,df_prod.ProductSubcategoryID==df_subcat.ProductSubcategoryID,"outer").count()
```

```python
# left join - pełna tabele z produktami
df_prod.join(df_subcat,df_prod.ProductSubcategoryID==df_subcat.ProductSubcategoryID,"left").show(5)

df_prod.join(df_subcat,df_prod.ProductSubcategoryID==df_subcat.ProductSubcategoryID,"left").count()
```

```python
# leftanti - produkty bez podkategorii
df_prod.join(df_subcat,df_prod.ProductSubcategoryID==df_subcat.ProductSubcategoryID,"leftanti").show(5)

df_prod.join(df_subcat,df_prod.ProductSubcategoryID==df_subcat.ProductSubcategoryID,"leftanti").count()
```

```python
# right join - pełna tabela z kategoriami
df_prod.join(df_subcat,df_prod.ProductSubcategoryID==df_subcat.ProductSubcategoryID,"right").show(5)

df_prod.join(df_subcat,df_prod.ProductSubcategoryID==df_subcat.ProductSubcategoryID,"right").count()
```
