```python
%%sql
-- używa "magic commands"
-- bardziej zwięzły kod, nie przypisuje wyniku bezpośrednio do ramki danych
-- wygodny dla zapytań ad-hoc
SELECT * FROM AdventureWorks.Production_Product LIMIT 5
```

```python
%%sql
DROP TABLE IF EXISTS products;

-- tworzenie tabeli
-- wspierane 2 rodzaje ograniczeń: NULL/ NOT NULL oraz CHECK  
CREATE TABLE products (
    ProductID int,
    ProductSubcategoryID int,
    Name string,
    Color string
) USING DELTA
```

```python
%%sql
-- wstawienie danych
INSERT INTO Products (ProductID, ProductSubcategoryID, Name, Color)
SELECT ProductID, ProductSubcategoryID, Name, Color FROM Production_Product
```

```python
%%sql
SELECT * FROM products LIMIT 1
```

```python
%%sql
SELECT Color, count(*)
FROM products
GROUP BY Color
```

```python
%%sql

SELECT *
FROM products AS p
JOIN Production_ProductSubcategory AS ps ON ps.ProductSubcategoryID = p.ProductSubcategoryID
JOIN Production_ProductCategory AS pc ON pc.ProductCategoryID = ps.ProductCategoryID 
LIMIT 5
```
